```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import cityblock

# --- OPERATIONS RESEARCH PRINCIPLE: MANHATTAN DISTANCE ---
# In a warehouse, we cannot walk through shelves (Euclidean distance).
# We use Manhattan Distance: |x1 - x2| + |y1 - y2| to simulate aisle navigation.

def calculate_optimized_route(start_node, locations):
    """
    Implements a 'Greedy Nearest Neighbor' algorithm to find an efficient
    picking path through a warehouse grid.
    """
    route = [start_node]
    unvisited = list(locations.values())
    total_distance = 0

    current_pos = start_node

    while unvisited:
        # Calculate Manhattan distances to all remaining items
        distances = [cityblock(current_pos, item) for item in unvisited]

        # Optimization: Select the mathematically 'nearest' neighbor
        closest_idx = np.argmin(distances)
        next_node = unvisited.pop(closest_idx)

        total_distance += distances[closest_idx]
        route.append(next_node)
        current_pos = next_node

    # Return to the packing station (Start)
    total_distance += cityblock(current_pos, start_node)
    route.append(start_node)

    return np.array(route), total_distance

# --- MOCK DATA: WAREHOUSE NODES (X, Y Coordinates) ---
# (0,0) represents the Packing/Shipping station
warehouse_items = {
    'Aisle 1/Shelf A': (2, 10),
    'Aisle 4/Shelf B': (8, 3),
    'Aisle 2/Shelf C': (4, 15),
    'Aisle 10/Shelf D': (20, 2),
    'Aisle 5/Shelf E': (10, 10)
}
start_point = (0, 0)

# Execute Optimization
optimized_path, travel_dist = calculate_optimized_route(start_point, warehouse_items)

# --- VISUALIZATION FOR PORTFOLIO ---
plt.figure(figsize=(10, 6))
plt.plot(optimized_path[:, 0], optimized_path[:, 1], '--o', color='blue', label='Picking Path')
plt.scatter(optimized_path[0,0], optimized_path[0,1], color='red', s=100, label='Start/End (Dock)')
plt.title(f"Warehouse Picker Route Optimization\nTotal Distance: {travel_dist} units")
plt.xlabel("Aisle Number (X)")
plt.ylabel("Shelf Position (Y)")
plt.grid(True)
plt.legend()
plt.show()

print(f"Mathematical Optimization Complete.")
print(f"The picker will travel {travel_dist} units to collect all items.")
```
